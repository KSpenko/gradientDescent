Definirajmo funkcijo polinoma, ki jo bomo minimizirali. Poljuben polinom najbolj intuitivno zapišemo z njegovimi ničlami. Igrali se bomo z dvema primeroma:
- kvadratna funkcija,
- bolj zapletena funkcija, s šestimi ničlami.

Več ničel kot ima polinoma, več minimumov imamo lahko. Seveda lahko sami definirate poljuben polinom in ga preizkusite.

```python
def polynomial(x):
    skalirniFaktor = 30.13
    # y = (x-0.5)**2
    y = (x-0.15)*(x-0.28)*(x-0.46)*(x-0.95)*(x-1.25)*(x-1.45)
    return skalirniFaktor * y
```

Najprej si oglejmo kako dobro poiscejo minimum metode iz strokovnih knjiznic.

```python
from scipy.optimize import golden, fmin_cg

print('fmin_cg = ', fmin_cg(polynomial, 1.5))
print('golden = ', golden(polynomial, brack=[0.2, 1.2]))
```
